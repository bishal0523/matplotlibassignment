```python
#

# q1.What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib:

# Matplotlib is a powerful and widely used plotting library in Python for creating static, interactive, and publication-quality visualizations.
# It is used for visualizing data in various formats such as line plots, scatter plots, bar plots, histograms, etc.
# Five plots that can be plotted using the Pyplot module of Matplotlib:
# Line Plot
# Scatter Plot
# Bar Plot
# Histogram
# Box Plot
# what is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot:

# A scatter plot is a type of plot that displays values for two variables as points on a Cartesian plane, where each point represents an observation in the data.
# It is used to visualize the relationship between two continuous variables.
# Here's the code to generate and plot a scatter plot using the provided data:
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

```

```python
#3
# Why is the subplot() function used? Draw four line plots using the subplot() function:
# The subplot() function is used to create multiple plots (subplots) within the same figure.
# It allows for better organization and comparison of multiple plots.
# Here's the code to draw four line plots using the subplot() function:
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Data for line plots
data = [
    (np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([50, 20, 40, 20, 60, 70])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150]))
]

# Plotting subplots
plt.figure(figsize=(10, 8))
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.plot(data[i][0], data[i][1])
    plt.title(f'Line Plot {i+1}')

plt.tight_layout()
plt.show()
```

```python
#What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot:
# A bar plot is a type of plot that presents categorical data with rectangular bars, where the length of each bar represents the value of the data.
# It is used to compare values across different categories and to visualize the distribution or frequency of categorical data.
# Here's the code to plot a bar plot and a horizontal bar plot using the provided data:
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(10, 5))

# Vertical bar plot
plt.subplot(1, 2, 1)
plt.bar(company, profit, color='skyblue')
plt.title('Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')

# Horizontal bar plot
plt.subplot(1, 2, 2)
plt.barh(company, profit, color='lightgreen')
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit')
plt.ylabel('Company')

plt.tight_layout()
plt.show()
```
